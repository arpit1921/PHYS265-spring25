$$
\frac{dy}{dt} = v
$$
Thus the equation comes out to be: 
$$
\frac{dv}{dt} = -g + \alpha v^\gamma
$$

```python
#Part(c) 
def dragproj(t,s,a,g0,gamma):
    y,v = s
    dydt = v
    dvdt = -g0+a*(v**gamma)
    return [dydt,dvdt] 

g0 = 9.81
a = 0 
gamma = 2 
t0 = 0 
tf = 30 
n = 101
t = np.linspace(t0,tf,n)
y0 = 4000 
v0 = 0
def reaching(t,s,a,g0,gamma):
    y,v = s
    return y
soln = scint.solve_ivp(fun = dragproj,
                          t_span=(t0,tf), # tuple
                          y0=[y0,v0], #as an arrray
                          args=(a,g0,gamma),
                          t_eval = t,events=reaching
                      )

y = soln.y[0] 
v = soln.y[1] 
tt = soln.t 
fig,ax = plt.subplots() 
ax1 = ax.twinx() 
p1 = ax.plot(tt,y,'orange',label='Position'); 
p2 = ax1.plot(tt,v,label='Velocity'); 
fig.legend();
ax.set_xlabel("Time(s)"); 
ax.set_ylabel("Position(m)"); 
ax1.set_ylabel("Velocity(m/s)");
ax.axhline(0.0,color='k',linestyle='--');
ax.set_title("Figure 1");
```

```python
#Part(a)
def dragproj_2(t,s,a,g0,gamma,Re,omega):
    yx,yy,vx,vy = s
    dyxdt = vx 
    dyydt = vy
    g = (g0*((Re+yy-4000)/Re)) 
    dvxdt = 2*omega*vy 
    dvydt = -g+a*(np.abs(vy)**gamma) - 2*omega*vx 
    return [dyxdt,dyydt,dvxdt,dvydt] 

omega  = 7.272e-5 
def reaching(t,s,a,g0,gamma,Re,omega):
    yx,yy,vx,vy = s
    return yx 
```

Part(b) 

As we know that: 
$$
\rho = \frac{Mass}{Volume} 
$$ 
This implits that: 
$$
\frac{\rho_M}{\rho_E} = \frac{\frac{M_M}{V_M}}{\frac{M_E}{V_E}} 
$$ 
=>  
$$ 
\frac{\rho_M}{\rho_E} = \frac{M_M}{M_E} * \frac{V_E}{V_M} 
$$ 
=> 
$$ 
\frac{\rho_M}{\rho_E} = \frac{M_M}{M_E} * \frac{4/3 \pi R_E^3}{4/3 \pi R_M^3} 
$$ 
=>
$$ 
\frac{\rho_M}{\rho_E} = \frac{M_M}{M_E} * \frac{R_E^3}{R_M^3} 
$$
Thus: 
$$
\frac{\rho_M}{\rho_E} = 0.6081591$$

                                    This implies that the Earth is around ~1.64 times more dense than the Moon 

As we know from above 
$$ 
v = \sqrt{\frac{GM}{R}} 
$$ 
And 
$$
T = \frac{2\pi R}{v} 
$$ 
This implies that:
$$
T = \frac{2\pi R}{\sqrt{\frac{GM}{R}} } 
$$ 
Which gives us the expression for T as: 
$$
T = 2\pi \sqrt{\frac{R^3}{GM}} 
$$  
Since : 
$$ 
M = \rho * V 
$$ 
Thus, 
$$ 
M = \rho * \frac{4}{3} \pi R^3 
$$ 
Which tells us that: 
$$
T = 2\pi \sqrt{\frac{R^3}{G\rho * \frac{4}{3} \pi R^3 }} 
$$  
And thus we have: 
$$
T =\sqrt{\frac{3\pi}{4G\rho}} 
$$  
Thus we can have the relation between Time Period and Density as: 
$$ 
T \propto \frac{1}{\sqrt{\rho}} 
$$

This can also be seen from the values as if we take the ratio between orbital time of Earth and Moon, 
$$ 
\frac{T_M}{T_E} = \frac{5065.08265064}{6498.89413048} 
$$ 
$$
\frac{T_M}{T_E} = 0.7793760829068774 
$$ 
And if we take the sqrt of the ratio of their density: 
$$
\sqrt{\frac{\rho_M}{\rho_E}} = 0.779845561 
$$ 
Both the values are approximately equal and thus proves the relation between density of a celestial body and its orbital time
